```lua
nData = 20
kWidth = 1
xTrain = torch.linspace(-1, 1, nData)
yTrain = torch.pow(xTrain, 2) + torch.mul(torch.randn(nData), 0.1)
```

```lua
Plot = require 'itorch.Plot'
plot = Plot():circle(xTrain,yTrain, 'red'):draw()
```

```lua
function phi(x,y)
    return torch.exp(-(1/kWidth)*torch.sum(torch.pow(x-y,2)))
end
```

```lua
Phi = torch.Tensor(nData, nData)
```

```lua
for i = 1, nData do
    for j = 1, nData do
        Phi[i][j] = phi(xTrain[{{i}}], xTrain[{{j}}])
    end
end
```
